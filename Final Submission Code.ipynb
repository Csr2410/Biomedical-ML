```python
import os
import numpy as np
import pandas as pd
import librosa
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter

# Paths
data_path = r'C:\Users\ncsek\Desktop\EE502final\Dataset4'
output_folder = r'C:\Users\ncsek\Desktop\EE502final\ProcessedFeatures'
os.makedirs(output_folder, exist_ok=True)

# Window settings
SAMPLE_RATE = 16000
WINDOW_SIZE = int(0.02 * SAMPLE_RATE)  # 20 ms
HOP_LENGTH = WINDOW_SIZE // 2

# Feature Extraction Function
def extract_features(y, sr):
    """
    Extracts multiple audio features and ensures consistent dimensionality.
    Handles edge cases where features might be empty or invalid.
    """
    try:
        # Extract features
        mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13, hop_length=HOP_LENGTH, n_fft=WINDOW_SIZE)
        chroma = librosa.feature.chroma_stft(y=y, sr=sr) if len(y) > 0 else np.zeros((12, 1))
        spectral_contrast = librosa.feature.spectral_contrast(y=y, sr=sr) if len(y) > 0 else np.zeros((7, 1))
        rmse = librosa.feature.rms(y=y) if len(y) > 0 else np.zeros((1, 1))
        zcr = librosa.feature.zero_crossing_rate(y=y) if len(y) > 0 else np.zeros((1, 1))

        # Ensure all features are at least 2D (even for single values)
        mfcc = mfcc if mfcc.ndim == 2 else np.zeros((13, 1))
        chroma = chroma if chroma.ndim == 2 else np.zeros((12, 1))
        spectral_contrast = spectral_contrast if spectral_contrast.ndim == 2 else np.zeros((7, 1))
        rmse = rmse if rmse.ndim == 2 else np.zeros((1, 1))
        zcr = zcr if zcr.ndim == 2 else np.zeros((1, 1))

        # Concatenate all features
        features = np.concatenate([
            np.mean(mfcc, axis=1), np.std(mfcc, axis=1),
            np.mean(chroma, axis=1), np.std(chroma, axis=1),
            np.mean(spectral_contrast, axis=1), np.std(spectral_contrast, axis=1),
            np.mean(rmse).reshape(1), np.std(rmse).reshape(1),
            np.mean(zcr).reshape(1), np.std(zcr).reshape(1)
        ])

    except Exception as e:
        # Handle any unexpected errors gracefully
        print(f"Error processing audio file with length {len(y)}: {e}")
        features = np.zeros(60)  # Default to zeros for consistency

    return features


# Process Data Function
def process_data(data_path, fraction=0.5):
    """
    Process a fraction of data from each class folder.
    """
    features, labels = [], []
    label_dirs = [d for d in os.listdir(data_path) if os.path.isdir(os.path.join(data_path, d))]
    for label in label_dirs:
        label_path = os.path.join(data_path, label)
        files = [f for f in os.listdir(label_path) if f.endswith(".wav")]
        selected_files = np.random.choice(files, size=int(len(files) * fraction), replace=False)
        for file in selected_files:
            file_path = os.path.join(label_path, file)
            y, sr = librosa.load(file_path, sr=SAMPLE_RATE)
            features.append(extract_features(y, sr))
            labels.append(label)
    return np.array(features), labels

# Save Extracted Features
def save_features_to_csv(features, labels, file_path):
    """
    Save extracted features and their corresponding labels to a CSV file.
    """
    df = pd.DataFrame(features)
    df['label'] = labels
    df.to_csv(file_path, index=False)
    print(f"Features saved to {file_path}")

# Save Train-Test Split
def save_train_test_split(X_train, X_test, y_train, y_test, output_folder):
    """
    Save training and testing datasets into separate CSV files.
    """
    os.makedirs(output_folder, exist_ok=True)
    
    # Save train data
    train_df = pd.DataFrame(X_train)
    train_df['label'] = y_train
    train_df.to_csv(os.path.join(output_folder, "train_data.csv"), index=False)
    
    # Save test data
    test_df = pd.DataFrame(X_test)
    test_df['label'] = y_test
    test_df.to_csv(os.path.join(output_folder, "test_data.csv"), index=False)
    
    print(f"Train and test data saved to {output_folder}")

# Save PCA-Transformed Features
def save_pca_features(X_train_pca, X_test_pca, y_train, y_test, output_folder):
    """
    Save PCA-transformed train and test datasets to separate CSV files.
    """
    os.makedirs(output_folder, exist_ok=True)
    
    # Save train PCA data
    train_df = pd.DataFrame(X_train_pca)
    train_df['label'] = y_train
    train_df.to_csv(os.path.join(output_folder, "train_pca_data.csv"), index=False)
    
    # Save test PCA data
    test_df = pd.DataFrame(X_test_pca)
    test_df['label'] = y_test
    test_df.to_csv(os.path.join(output_folder, "test_pca_data.csv"), index=False)
    
    print(f"PCA-transformed data saved to {output_folder}")

# Process Data and Save Features
X, y = process_data(data_path, fraction=0.5)
save_features_to_csv(X, y, os.path.join(output_folder, "features.csv"))

# Encode Labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split Data and Save Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, stratify=y_encoded, random_state=42)
save_train_test_split(X_train, X_test, y_train, y_test, os.path.join(output_folder, "train_test_split"))

# Standardize Features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Perform PCA and Save PCA Features
pca = PCA(n_components=0.95)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
save_pca_features(X_train_pca, X_test_pca, y_train, y_test, os.path.join(output_folder, "pca_data"))

# Train and Evaluate Models
models = {
    "Random Forest": RandomForestClassifier(random_state=42),
    "SVM (without PCA)": SVC(random_state=42),
    "SVM (with PCA)": SVC(random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42)
}

results = {}
for name, model in models.items():
    if "PCA" in name:
        model.fit(X_train_pca, y_train)
        y_pred = model.predict(X_test_pca)
    else:
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
    
    accuracy = accuracy_score(y_test, y_pred)
    results[name] = accuracy
    print(f"\n{name} Accuracy: {accuracy:.4f}")
    print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))

# Plot the Comparison
plt.figure(figsize=(10, 6))
sns.barplot(x=list(results.keys()), y=list(results.values()))
plt.title("Model Comparison")
plt.ylabel("Accuracy")
plt.xlabel("Model")
plt.xticks(rotation=45)
plt.show()
```

```python
# Full Pipeline for Denoised Data

import os
import numpy as np
import pandas as pd
import librosa
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
import soundfile as sf

# Paths
data_path = r'C:\Users\ncsek\Desktop\EE502final\Dataset4'
denoised_data_path = r'C:\Users\ncsek\Desktop\EE502final\Processed4\DenoisedData'
output_folder_denoised = r'C:\Users\ncsek\Desktop\EE502final\Processed4\DenoisedFeatures'
os.makedirs(output_folder_denoised, exist_ok=True)

# Window settings
SAMPLE_RATE = 16000
WINDOW_SIZE = int(0.02 * SAMPLE_RATE)
HOP_LENGTH = WINDOW_SIZE // 2

# Denoising Function
def denoise_audio(input_path, output_path, fraction=0.5):
    """
    Perform denoising on audio files and save the processed files.
    """
    for label in os.listdir(input_path):
        label_path = os.path.join(input_path, label)
        if os.path.isdir(label_path):
            output_label_path = os.path.join(output_path, label)
            os.makedirs(output_label_path, exist_ok=True)
            files = [f for f in os.listdir(label_path) if f.endswith(".wav")]
            selected_files = np.random.choice(files, size=int(len(files) * fraction), replace=False)
            for file in selected_files:
                file_path = os.path.join(label_path, file)
                y, sr = librosa.load(file_path, sr=SAMPLE_RATE)
                y_denoised = librosa.effects.trim(y, top_db=20)[0]
                sf.write(os.path.join(output_label_path, file), y_denoised, sr)

```

```python
# Perform Denoising
denoise_audio(data_path, denoised_data_path, fraction=0.5)
print(f"Denoised audio files saved in {denoised_data_path}")

# Feature Extraction Function
def extract_features(y, sr):
    """
    Extract audio features from a given audio signal.
    Handles edge cases where features might be empty or invalid.
    """
    try:
        # Extract features
        mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13, hop_length=HOP_LENGTH, n_fft=WINDOW_SIZE)
        chroma = librosa.feature.chroma_stft(y=y, sr=sr) if len(y) > 0 else np.zeros((12, 1))
        spectral_contrast = librosa.feature.spectral_contrast(y=y, sr=sr) if len(y) > 0 else np.zeros((7, 1))
        rmse = librosa.feature.rms(y=y) if len(y) > 0 else np.zeros((1, 1))
        zcr = librosa.feature.zero_crossing_rate(y=y) if len(y) > 0 else np.zeros((1, 1))

        # Ensure all features have consistent dimensions
        mfcc = mfcc if mfcc.ndim == 2 else np.zeros((13, 1))
        chroma = chroma if chroma.ndim == 2 else np.zeros((12, 1))
        spectral_contrast = spectral_contrast if spectral_contrast.ndim == 2 else np.zeros((7, 1))
        rmse = rmse if rmse.ndim == 2 else np.zeros((1, 1))
        zcr = zcr if zcr.ndim == 2 else np.zeros((1, 1))

        # Combine features
        features = np.concatenate([
            np.mean(mfcc, axis=1), np.std(mfcc, axis=1),
            np.mean(chroma, axis=1), np.std(chroma, axis=1),
            np.mean(spectral_contrast, axis=1), np.std(spectral_contrast, axis=1),
            np.mean(rmse).reshape(1), np.std(rmse).reshape(1),
            np.mean(zcr).reshape(1), np.std(zcr).reshape(1)
        ])

    except Exception as e:
        print(f"Error processing audio sample: {e}")
        # Default to zero vector if there is an issue
        features = np.zeros(60)  # Adjust size based on total number of features

    return features


# Process Denoised Data
def process_data(data_path, fraction=0.5):
    features, labels = [], []
    label_dirs = [d for d in os.listdir(data_path) if os.path.isdir(os.path.join(data_path, d))]
    for label in label_dirs:
        label_path = os.path.join(data_path, label)
        files = [f for f in os.listdir(label_path) if f.endswith(".wav")]
        selected_files = np.random.choice(files, size=int(len(files) * fraction), replace=False)
        for file in selected_files:
            file_path = os.path.join(label_path, file)
            y, sr = librosa.load(file_path, sr=SAMPLE_RATE)
            features.append(extract_features(y, sr))
            labels.append(label)
    return np.array(features), labels

# Process and Save Features
X_denoised, y_denoised = process_data(denoised_data_path, fraction=0.5)

# Save Features
def save_features_to_csv(features, labels, file_path):
    df = pd.DataFrame(features)
    df['label'] = labels
    df.to_csv(file_path, index=False)
    print(f"Features saved to {file_path}")

save_features_to_csv(X_denoised, y_denoised, os.path.join(output_folder_denoised, "features.csv"))

# Encode Labels
label_encoder = LabelEncoder()
y_denoised_encoded = label_encoder.fit_transform(y_denoised)

# Split Data
X_train_denoised, X_test_denoised, y_train_denoised, y_test_denoised = train_test_split(
    X_denoised, y_denoised_encoded, test_size=0.2, stratify=y_denoised_encoded, random_state=42
)

# Standardize Features
scaler = StandardScaler()
X_train_denoised_scaled = scaler.fit_transform(X_train_denoised)
X_test_denoised_scaled = scaler.transform(X_test_denoised)

# Perform PCA
pca = PCA(n_components=0.95)
X_train_denoised_pca = pca.fit_transform(X_train_denoised_scaled)
X_test_denoised_pca = pca.transform(X_test_denoised_scaled)

# Models
models = {
    "Random Forest": RandomForestClassifier(random_state=42),
    "SVM (without PCA)": SVC(random_state=42),
    "SVM (with PCA)": SVC(random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42)
}

# Cross-Validation and Model Training
cv_results_denoised = {}
for name, model in models.items():
    print(f"\nPerforming Cross-Validation for {name} on Denoised Data")
    if "PCA" in name:
        scores = cross_val_score(model, X_train_denoised_pca, y_train_denoised, cv=5, scoring="accuracy")
    else:
        scores = cross_val_score(model, X_train_denoised_scaled, y_train_denoised, cv=5, scoring="accuracy")
    cv_results_denoised[name] = (scores.mean(), scores.std())
    print(f"{name} Cross-Validation Mean Accuracy = {scores.mean():.4f}, Std Dev = {scores.std():.4f}")

# Statistical Testing
compare_model_performances(
    cv_results_denoised["Random Forest"], cv_results_denoised["SVM (without PCA)"],
    "Random Forest", "SVM (without PCA)"
)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

def plot_confusion_matrix(y_true, y_pred, class_labels, model_name):
    """
    Plot the confusion matrix for a given model's predictions.
    """
    cm = confusion_matrix(y_true, y_pred, labels=np.arange(len(class_labels)))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_labels)
    disp.plot(cmap=plt.cm.Blues, xticks_rotation=45)
    plt.title(f"{model_name} Confusion Matrix")
    plt.show()

```

```python
# Evaluate and Plot Results
results_denoised = {}
for name, model in models.items():
    print(f"\nTraining and Evaluating {name} on Denoised Data")
    if "PCA" in name:
        model.fit(X_train_denoised_pca, y_train_denoised)
        y_pred_denoised = model.predict(X_test_denoised_pca)
    else:
        model.fit(X_train_denoised_scaled, y_train_denoised)
        y_pred_denoised = model.predict(X_test_denoised_scaled)
    
    # Calculate Accuracy
    accuracy = accuracy_score(y_test_denoised, y_pred_denoised)
    results_denoised[name] = accuracy
    print(f"{name} Accuracy on Denoised Data: {accuracy:.4f}")

    # Ensure Matching Classes
    unique_classes_in_test = np.unique(y_test_denoised)

    # Generate Classification Report
    print(classification_report(
        y_test_denoised,
        y_pred_denoised,
        labels=unique_classes_in_test,
        target_names=[label_encoder.classes_[i] for i in unique_classes_in_test]
    ))

    # Plot Confusion Matrix
    plot_confusion_matrix(
        y_test_denoised,
        y_pred_denoised,
        [label_encoder.classes_[i] for i in unique_classes_in_test],
        f"{name} on Denoised Data"
    )

```
